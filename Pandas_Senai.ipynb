```python
#@title Renomeando Colunas:
df.rename(columns={"address":"endereco"}, inplace=True)
```

```python
grafico = df.groupby("tipo")[["total"]].mean().sort_values("total", ascending=False)
```

###**Montando Gr√°fico de dispers√£o com MatPlotLib** üìä

```python
import matplotlib.pyplot as plt
```

```python
#O gr√°fico de dispers√£o analisa a rela√ß√£o entre duas vari√°veis
fig = plt.figure(figsize=(8,6))
eixo = fig.add_axes([0,0,1,1])
eixo.set_title("Total por tipo de moradia")
eixo.set_xlabel("Total Mensal")
eixo.set_ylabel("Tipo Moradia")
eixo.legend(["Total Mensal"], loc="upper right")
eixo.scatter(df["total"], df["tipo"], color="red")
```

```python
fig = plt.figure(figsize=(8,6))
eixo = fig.add_axes([0,0,1,1])
eixo.set_title("Rela√ß√£o √Årea x Aluguel")
eixo.set_xlabel("Total Mensal")
eixo.set_ylabel("√Årea")
eixo.legend(["Total Mensal"], loc="upper right")
eixo.scatter(df["total"], df["area"], color="orange", marker="_")
```
